```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df_predictors = pd.read_excel('/content/Extended Data_predictors.xlsx')
print("Предикторы - форма:", df_predictors.shape)
print(df_predictors.head())

df_target = pd.read_excel('/content/Extended Data_response.xlsx')
print("\nЦелевая переменная - форма:", df_target.shape)
print(df_target.head())

assert len(df_predictors) == len(df_target), "Количество строк в файлах не совпадает!"


df_full = df_predictors.copy()
df_full['Kc'] = df_target['Kc']  # предполагая, что колонка называется 'K_IC'

print(f"\nИтоговый датасет: {df_full.shape[0]} строк, {df_full.shape[1]} колонок")
print(f"Колонки: {df_full.columns.tolist()}")

# Проверка на пропуски
print(f"\nПропуски в данных:\n{df_full.isnull().sum()}")

# ------------------------------------------------------------
# Этап 2. Разделение данных на обучающую и тестовую выборки
# ------------------------------------------------------------

# Разделение 80/20 с перемешиванием
train_df, test_df = train_test_split(df_full, test_size=0.2, random_state=42)

print(f"\n{'='*50}")
print(f"Обучающая выборка: {len(train_df)} строк (80%)")
print(f"Тестовая выборка: {len(test_df)} строк (20%)")
print(f"{'='*50}")

# Отделяем предикторы (X) от целевой переменной (y)
X_train = train_df[['P', 'a', 'c']].values  # убедитесь, что колонки называются именно P, a, c
y_train = train_df['Kc'].values

X_test = test_df[['P', 'a', 'c']].values
y_test = test_df['Kc'].values

# ------------------------------------------------------------
# (Опционально) Статистика по разделению
# ------------------------------------------------------------

print("\nСтатистика обучающей выборки:")
print(f"P: min={train_df['P'].min():.4f}, max={train_df['P'].max():.4f}, mean={train_df['P'].mean():.4f}")
print(f"a: min={train_df['a'].min():.4f}, max={train_df['a'].max():.4f}, mean={train_df['a'].mean():.4f}")
print(f"c: min={train_df['c'].min():.4f}, max={train_df['c'].max():.4f}, mean={train_df['c'].mean():.4f}")
print(f"K_IC: min={train_df['Kc'].min():.4f}, max={train_df['Kc'].max():.4f}, mean={train_df['Kc'].mean():.4f}")

print("\nСтатистика тестовой выборки:")
print(f"P: min={test_df['P'].min():.4f}, max={test_df['P'].max():.4f}, mean={test_df['P'].mean():.4f}")
print(f"a: min={test_df['a'].min():.4f}, max={test_df['a'].max():.4f}, mean={test_df['a'].mean():.4f}")
print(f"c: min={test_df['c'].min():.4f}, max={test_df['c'].max():.4f}, mean={test_df['c'].mean():.4f}")
print(f"K_IC: min={test_df['Kc'].min():.4f}, max={test_df['Kc'].max():.4f}, mean={test_df['Kc'].mean():.4f}")

# ------------------------------------------------------------
# Сохранение результатов (опционально)
# ------------------------------------------------------------

# Сохранить разделённые выборки в Excel или CSV
# train_df.to_excel('обучающая_выборка.xlsx', index=False)
# test_df.to_excel('тестовая_выборка.xlsx', index=False)

print("\nДанные готовы для обучения модели!")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```
