```python
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import to_categorical
import keras
```

```python

alph = ['A','Р‘','Р’','Р“','Р”','Р•','Р–','Р—'] # РЅР°Р±РѕСЂ Р±СѓРєРІ

# Р—Р°РіСЂСѓР¶Р°РµРј РёР·РѕР±СЂР°Р¶РµРЅРёСЏ 5Г—7 РїРёРєСЃРµР»РµР№
img = []
for a in alph:
    im = cv2.imread('/content/' + a + '.png', cv2.IMREAD_GRAYSCALE)
    img.append(im)

# Р“РѕС‚РѕРІРёРј X (РІС…РѕРґС‹)
X = []
for im in img:
    vec = im.reshape(-1)          # РІРµРєС‚РѕСЂРёР·Р°С†РёСЏ
    vec = np.where(vec == 0, 1, 0)
    X.append(vec)

X = np.array(X)

# Р“РѕС‚РѕРІРёРј y (РјРµС‚РєРё)
y_int = np.arange(len(alph))
y = to_categorical(y_int, num_classes=8)

# РЎРѕР·РґР°С‘Рј РјРѕРґРµР»СЊ softmax-РєР»Р°СЃСЃРёС„РёРєР°С†РёРё
model = Sequential()
model.add(Dense(64, input_shape=(X.shape[1],), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(8, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# РћР±СѓС‡РµРЅРёРµ
model.fit(X, y, epochs=300, verbose=1)

```

```python
def predict_letter(path):
    im = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    print(im)
    vec = im.reshape(-1)
    vec = np.where(vec == 0, 1, 0)
    vec = vec.reshape(1, -1)

    pred = model.predict(vec)
    idx = np.argmax(pred)

    return alph[idx]

# Р‘СѓРєРІР° Рђ
letter = predict_letter('/content/A.png')
print("Р‘СѓРєРІР°:", letter)

# Р‘СѓРєРІР° Р“
letter = predict_letter('/content/Р“.png')
print("Р‘СѓРєРІР°:", letter)

# Р‘СѓРєРІР° Р–
letter = predict_letter('/content/Р–.png')
print("Р‘СѓРєРІР°:", letter)

# Р‘СѓРєРІР° Р“ СЃ С€СѓРјРѕРј
letter = predict_letter('/content/Р“_РЁРЈРњ.png')
print("Р‘СѓРєРІР°:", letter)

# Р‘СѓРєРІР° Рђ СЃ С€СѓРјРѕРј
letter = predict_letter('/content/Рђ_РЁРЈРњ.png')
print("Р‘СѓРєРІР°:", letter)

# Р‘СѓРєРІР° Р” СЃ С€СѓРјРѕРј
letter = predict_letter('/content/Р”_РЁРЈРњ.png')
print("Р‘СѓРєРІР°:", letter)
```
